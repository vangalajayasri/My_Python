## 3. How is the distribution of applications by their size types in the dataset?

```python
# checking Size of the apps
df['Size'].unique()
```

```python
df['Size'].isnull().sum()
```

```python
# replace commas with empty string
df['Size'] = df['Size'].str.replace(',', '')
df['Size'].unique()
```

```python
# df['Size'].unique().sort_values()
```

```python
# define a function to convert a single value to a number
def convert_value(value):
    # check if the value is already a float
    if isinstance(value, float):
        return value
    
    # check if the value ends with "M"
    if value.endswith('M'):
        # remove the "M" character and return the value of MB as float
        return float(value[:-1])
    # check if the value ends with "k"
    elif value.endswith('k'):
        
        return float(value[:-1]) / 1024
    # if the value is 'Varies with device' return 0
    elif value.endswith('device'):
        return 0
    elif value.endswith('G'):
        
        return float(value[:-1]) * 1024
    else:
        return float(value)

# apply the function to the "Size" column
df['Size'] = df['Size'].apply(convert_value)
print(df['Size'])
df['Size'].unique()
```

```python
df['Size']=df['Size'].astype(float)
Size=df['Size']
Size
# define a function to create 5 categories of size type
def size_cat(size):
    if size == 0.0:
        return 'Size not provided'
    elif size > 0.0 and size <= 100.0:
        return 'Under 100 MB'
    elif size > 100.0 and size <= 500.0:
        return 'Between 100 and 500 MB'
    elif size > 500.0 and size <= 1024.0:
        return 'Between 500 MB and 1 GB '
    else:
        return 'More than 1 GB'
df['Size_Type']= df["Size"].apply(size_cat)
print("Unique values in 'Size Type':", df['Size_Type'].unique())
```

```python
# Print count of each category in 'Size Type'
print("Count of each category in 'Size Type':")
print(df['Size_Type'].value_counts())
```

## 12. category and calculate mean rating count for each category

```python
# Group by category and calculate mean rating count for each category
average_rating_count_by_category = df.groupby('Category')['Rating Count'].mean().sort_values(ascending=False).head(5)

# Plot multiline graph
plt.figure(figsize=(10, 6))
for category in average_rating_count_by_category.index:
    ratings = df[df['Category'] == category]['Rating Count']  
    plt.plot(ratings, label=category)

plt.xlabel('Rating Count')
plt.ylabel('Rating')
plt.title('Average Rating by Category')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

# 13.Correlation Heatmap of Google Play Store Data

```python
numerical_columns = ['Rating', 'Price', 'Size', 'Installs', 'Maximum Installs', 'Minimum Installs']
# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()
# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Google Play Store Data')
plt.show()
```

# 14.Category wise Histogram of Rating Counts

```python
# Group by category and calculate mean rating count for each category
average_rating_count_by_category = df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(5)
#print(average_rating_count_by_category)
# Plot histogram
plt.figure(figsize=(10, 6))
for category in average_rating_count_by_category.index:
    rating_counts = df[df['Category'] == category]['Rating']  # Get rating counts for the current category
    plt.hist(rating_counts, bins=20, alpha=0.5, label=category)  # Adjust the number of bins as needed

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Rating Counts by Category')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```
